# Trying stuff out

```julia
using Plots
using FFTW

sample_interval = 0.01

f = x -> sin(2*x) + 0.5sin(2*3x) + 0.3sin(2*5x)

t = 0:sample_interval:10
y = f.(t)

plt = plot(t, y, label="Original", lw=2)
```

```julia
noise = 0.5*randn(length(t))
y_noisy = y + noise

noise_overlayed_plt = plot(plt, t, y_noisy, label="Noisy", alpha=0.5)
```

```julia
power_original = abs.(fft(y)) ./ length(y)
power_noisy = abs.(fft(y_noisy)) ./ length(y)

frequencies = range(0, stop=1/(sample_interval), length=length(y))
x_lims = (0, 10)
x_ticks = 0:1:10

power_original_plt = plot(frequencies, power_original, label="Original", lw=2, xlims=x_lims, xticks=x_ticks)
power_noisy_plt = plot(frequencies, power_noisy, label="Noisy", lw=2, xlims=x_lims, xticks=x_ticks)

plot(noise_overlayed_plt, power_original_plt, power_noisy_plt,
    layout=grid(3,1),
    xlabel="Frequency",
    ylabel="Power",
    title="Power Spectrum")
```
